# 6. Principal component analysis (PCA) with and without z-score normalization: project the data to the first two principal components. Visualize the result as a scatter plot. 

```python
# PDA without z-score normalization:
m = X.shape[0]

# Center data:
Z = X - np.mean(X, axis=0)
# Compute covariance matrix:
C = Z.T@Z/(m-1)
# Eigendecompose covariance matrix:
evals, evecs = np.linalg.eigh(C)
# sort eigenvalues in decreasing order
idx = np.argsort(evals)[::-1]
evecs = evecs[:,idx]
#pick eigenvectors corresponding to 2 largest eigenvalues
M = evecs[:,:2].T
# Project your data:
Z_mapped = (M@Z.T).T
```

```python
# PDA with z-score normalization:
from scipy.stats import zscore
 
# z-score standardization
Zs = zscore(X, axis=0)

# Compute covariance matrix:
Cs = Zs.T@Zs/(m-1)
# Eigendecompose covariance matrix:
evalss, evecss = np.linalg.eigh(Cs)
# sort eigenvalues in decreasing order
idxs = np.argsort(evalss)[::-1]
evecss = evecss[:,idxs]
#pick eigenvectors corresponding to 2 largest eigenvalues
Ms = evecss[:,:2].T
# Project your data:
Zs_mapped = (Ms@Zs.T).T
```

# What is the proportion of variance explained in projections? What can be observed?

```python
# without z-score normalization / eigenvalues
evals = evals[idx]
sumevals= sum(evals)
#proportion of variance
print('proportion of variance (without z-score normalization)', (evals[0]+ evals[1])/sumevals, (evals[0])/sumevals, (evals[1])/sumevals)

# without z-score normalization / eigenvalues
evalss = evalss[idxs]
sumevalss= sum(evalss)
#proportion of variance
print('proportion of variance (with z-score normalization)', (evalss[0]+ evalss[1])/sumevalss)

```

Without z-score normalization, the proportion of variance is 0.978, which is a little bit higher (better) than with z_score normalization (0.958). Z-score normalization would have been essential, if e.g different lengths had been in different scale (m/cm/mm). If in such situation no standardization is carried out, the attribute with the
largest variance could dominate the first principal component.
